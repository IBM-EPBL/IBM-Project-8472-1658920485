## outlier removal- IQR method

```python
Q1= df.BMI.quantile(0.25)
Q3=df.BMI.quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
upper_limit =Q3 + 1.5*IQR
lower_limit =Q1 - 1.5*IQR
```

```python
upper_limit
```

```python
df=df[df.BMI<upper_limit]
```

```python
sns.boxplot(df.BMI)
```

```python
df.shape
```
